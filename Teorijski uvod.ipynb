## ResNet mreže

Rezidualne mreže, skraćeno ResNet, napravile su prekretnicu u dubokom učenju, a  prvi put su predstavljene 2015. godine u [radu](https://arxiv.org/pdf/1512.03385.pdf) koji je odneo pobedu na Imagenet takmičenju te godine. Do tog trenutka VGG19, sa 19 slojeva, bila je najkorišćenija mreža u dubokom učenju jer su se prilikom dodavanja novih slojeva dobijali sve lošiji rezultati. Primer toga vidimo na slici ispod gde se može uočiti da dublja mreža ima veću grešku i na trening i na test skupu. Razlog ovakvog ponašanja mreža sa više slojeva je uzrokovan problemom nestajućih gradijenata.

ResNet mreže izazvale su veliko iznenađenje jer su davale odlične rezultate na jako dubokim mrežama od 34, 50, 101 ili 152 sloja. Ove mreže ne samo što su imale manje greške od do tada korišćenih mreža, već su imale i manji broj parametara, što je olakšavalo trening. Razlog ovome su skip konekcije koje su ugrađene u arhitekturu ResNeta. Na slici ispod nalazi se jedan rezidualni blok, odnosno blok koji sadrži skip konekciju. Kao što se može videti, skip konekcije preskaču neke slojeve i izlazu rezidualnog bloka pridodaju njegov ulaz. Rezultat ovoga je očuvanje gradijenta tokom propagacije unazad, što omogućava da ne dođe do problema nestajućih gradijenata. ResNet mreže se sastoje iz više rezidualnih blokova koji su nadovezani jedan na drugi.

Prilikom treniranja mreže minimizujemo funkciju greške koja postaje sve kompleksnija kako dubina mreže raste. Na slici ispod vidimo zašto skip konekcije predstavljaju neophodni deo arhitekture veoma dubokih mreža. Na levoj slici prikazana je funkcija greške ResNet56 mreže koja nema skip konekcije, dok su na slici desno one dodate.

U prvom redu na slici nalaze se greške mreža koje imaju skip konekcije, a u drugom redu funkcije greške bez skip konekcija. 

Dodavanjem ovih konekcija pokazano je da sada dublje mreže daju bolje rezultate. Leva slika prikazuje greške na treningu(boldovane linije) i na testu(tanka linija) mreža sa 18 i 34 sloja  koje nemaju skip konekciju. Na slici desno su dodate skip konekcije. Jasno se vidi da u drugom slučaju sa povećanjem dubine mreže dolazi do poboljšanja njenogn rada.

Iako sa povećanjem broja slojeva dolazi do boljih rezultata, treba imati u vidu da ovo ne važi uvek i da prilikom korišćenja velikog broja slojeva dolazi do preprilagođavanja. Ovaj problem se može videti na slici ispod gde je greška između trening i test skupa vidno veća na mreži sa 1202 sloja u odnosu na mrežu sa 110 slojeva.
