## ResNet mreže

Rezidualne mreže, skraćeno ResNet,napravile su prekretnicu u dubokom učenju,a  prvi put su predstavljene 2015. godine u [radu](https://arxiv.org/pdf/1512.03385.pdf) koji je odneo pobedu na Imagenet takmičenju te godine. Do tog trenutka VGG19, sa 19 slojeva, bila je najkorišćenija mreža u dubokom učenju jer su se prilikom dodavanja novih slojeva dobijali sve lošiji rezultati. Primer toga vidimo na slici ispod gde se može uočiti da dublja mreža ima veću grešku i na trening i na test skupu. Razlog ovakvog ponašanja mreža sa više slojeva je problem nestajućih gradijenata.

ResNet mreže izazvale su veliko iznenađenje jer su davale odlične rezultate na jako dubokim mrežama od 34, 50, 101 i 152 sloja. Ove mreže ne samo što su imale manje greške od do tada korišćenih mreža, već su imale i manji broj parametara, što je olakšavao trening. Razlog ovome su skip konekcije koje su ugrađene u arhitekturu ResNeta. Na slici ispod nalazi se jedan rezidualni blok, odnosno blok koji sadrži skip konekciju.

Na levoj slici ispod prikazane su greske na trening (boldovana linija) i test skupu (isprekidana linija) na mrežama sa 18 i 34 sloja koje nemaju skip konekciju, dok je sa desne strane prikazana greška na mrežama kojima je ova konekcija dodata. Jasno se vidi da u drugom slučaju sa povećanjem dubine mreže dolazi do poboljšanja njenogn rada.

Iako sa povećanjem broja slojeva dolazi do boljih rezultata, treba imati u vidu da ovo ne važi uvek i da prilikom korišćenja velikog broja slojeva dolazi do preprilagođavanja. Ovaj problem se može videti na slici ispod gde je greska između trening i test skupa vidno veća na mreži sa 1202 slojeva u odnosu na mrežu sa 110 slojeva.
